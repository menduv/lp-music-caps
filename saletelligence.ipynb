### Columns and their meanings

- <span style="color: green;">'Health_Service_Area', 'County', 'Operating_Certificate_Number', 'Facility_ID', 'Facility_Name'</span>: Information about the Hospital/Clinic
- <span style="color: green;">'Age_Group', 'Zip_Code', 'Gender', 'Race', 'Ethnicity'</span>: Demographic information
- <span style="color: green;">'Length_Of_Stay'</span>: How long the patient stayed
- <span style="color: green;">'Type_Of_Admission'</span>: Self-explanatory 
- <span style="color: green;">'Patient_Disposition'</span>: Where did the patient go after the hospital treatment
- <span style="color: green;">'Diagnosis_Code', 'Diagnosis_Description'</span>: Linked columns; what is the diagnosis
- <span style="color: green;">'Procedure_Code', 'Procedure_Description'</span>: Linked column; what procedure was done on the patient
- <span style="color: green;">'APR_DRG_Code', 'APR_DRG_Description'</span>: Linked columns; All Patient Refined Diagnosis Related Groups (APR-DRGs) are a patient classification system that categorizes hospital inpatient stays based on clinical characteristics, severity of illness (SOI), and risk of mortality (ROM), used by payers and providers to determine payments for inpatient hospital services. 
- <span style="color: green;">'APR_MDC_Code', 'APR_MDC_Description'</span>: Linked columns; Major Diagnostic Category (MDC) is a classification system
- <span style="color: green;">'APR_Severity_Of_Illness_Code', 'APR_Severity_Of_Illness_Description', 'APR_Risk_Of_Mortality'</span>: Self-explanatory
- <span style="color: green;">'APR_Medical_Surgical_Description'</span>: Either a medical visit or a surgical visit
- <span style="color: green;">'Payment_Typology_1', 'Payment_Typology_2', 'Payment_Typology_3'</span>: Sources of payment
- <span style="color: green;">'Birth_Weight'</span>: Self-explanatory
- <span style="color: green;">'Is_Emergency_Department_Indicator'</span>: Redundant boolean column similar to 'Type_Of_Admission' column
- <span style="color: green;">'Total_Charges', 'Total_Cost'</span>: Self-Explanatory
